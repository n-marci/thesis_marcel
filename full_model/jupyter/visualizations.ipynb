# Some pretty graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
irrigation_yearly_power_usage = 1.943e+7
illumination_yearly_power_usage = 4.444e+7
solar_yearly_power_production = 2.8548e+8
print(irrigation_yearly_power_usage)
print(illumination_yearly_power_usage)
print(solar_yearly_power_production)
```

```python
tower_length = 55.17 #m - taken from 3d model
tower_width = 21.97 #m
tower_roof_area = tower_length * tower_width
print(f"roof area for one ot the towers: {tower_roof_area} m²")
```

```python
fig, ax = plt.subplots()
ax.bar(["Pump", "LEDs", "Solar"], [irrigation_yearly_power_usage, illumination_yearly_power_usage, solar_yearly_power_production], color=['blue', 'orange', 'green'])
ax.set_ylabel('Energy used / generated in Wh')
ax.set_title('Energy balance')
```
