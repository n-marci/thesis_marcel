```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_csv_data(csv_file, x_column, y_column, title, x_label, y_label, x_range=None, y_range=None):
    # Read CSV file
    df = pd.read_csv(csv_file)
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(df[x_column]-140, df[y_column])
    
    # Set title and labels
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    
    # Set data ranges if specified
    if x_range:
        plt.xlim(x_range)
    if y_range:
        plt.ylim(y_range)
    
    # Display the plot
    plt.grid(True)
    plt.show()

# Example usage
plot_csv_data(
    csv_file='exportedVariables.csv',
    x_column='time',
    y_column='gain.y',
    title='My CSV Data Graph',
    x_label='Date',
    y_label='Value',
    # x_range=(140, 200),
    # y_range=(0, 20)
)
```
